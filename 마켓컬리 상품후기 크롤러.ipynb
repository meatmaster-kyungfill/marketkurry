```python
import urllib.request
import urllib.parse
from bs4 import BeautifulSoup
import time
import re
import pandas as pd

pageNum = 2
count = 1
i = input('몇페이지 크롤링 할까요? : ')
lastPage = int(i)

result = []
while pageNum < lastPage + 1:
    url = f'https://www.kurly.com/shop/goods/goods_review_list.php?goodsno=29344&page={pageNum}'
    html = urllib.request.urlopen(url).read()
    soup = BeautifulSoup(html, 'html.parser')
    allcontent = soup.findAll('div', {'class':'tr_line'})
    
    for i in allcontent:
        product = i.find('div', {'class':'name_purchase'}).get_text()
        detail_product = i.find('p', {'class':'package'}).get_text()
        title = i.find('td', {'class':'subject'}).get_text()
        content = i.find('div',{'class':'inner_review'})
        content = content.get_text().replace("\n","@")
        content_review = re.findall('@@@.+@', content)
        grade = i.find('td', {'class':'user_grade grade_comm'}).get_text()
        date = i.find('td', {'class':'time'}).get_text()
        review_like_cnt = i.find('span', {'class':'review-like-cnt'}).get_text()
        review_hit_cnt = i.find('span', {'class':'review-hit-cnt'}).get_text()
        data = [product, detail_product, title, content_review, grade, date, review_like_cnt, review_hit_cnt]
        result.append(data)
        print(data)
    pageNum += 1
    count += 1
    time.sleep(1)
```

```python
import pandas as pd
master_data = pd.DataFrame(result)
master_data.columns = ['product', 'detail_product', 'title', 'content_review', 'grade', 'date', 'review-like-cnt', 'review-hit-cnt']
master_data
```

```python
master_data.to_excel('마켓컬리 니코하몽1.xlsx')
```
